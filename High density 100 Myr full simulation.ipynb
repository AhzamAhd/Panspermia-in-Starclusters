```python
import matplotlib.pyplot as plt
import numpy as np

# Define tolerances
tolerance_distance = 30 
tolerance_time = 0.5 

# Assume you have 10 dataframes: df1, df2, ..., df10
dataframes = [df1, df2, df3, df4, df5, df6, df7, df8, df9, df10]

# Create main axes
fig = plt.figure(figsize=(12, 8))
ax_main = fig.add_axes([0.1, 0.1, 0.65, 0.65])  # [left, bottom, width, height]
ax_histx = fig.add_axes([0.1, 0.78, 0.65, 0.2])  # Removed sharex=ax_main

# Collect data for histograms
all_times = []
all_distances = []

for df in dataframes:
    # Filter out rows where close encounter distance is less than 1000
    filtered_data = df[df['close_encounter_distance']*206265 < 970]
    
    # Further filter out rows where close encounter time is around 10
    filtered_data = filtered_data[(filtered_data['close_encounter_time'] < 10 - tolerance_time) | 
                                  (filtered_data['close_encounter_time'] > 10 + tolerance_time)]
    
    ax_main.scatter(filtered_data['close_encounter_time'], filtered_data['close_encounter_distance']*206265, color='blue')
    
    all_times.extend(filtered_data['close_encounter_time'].tolist())
    all_distances.extend((filtered_data['close_encounter_distance']*206265).tolist())

ax_main.set_yscale('log')
ax_main.set_xlabel('Close Encounter Time (Myr)')
ax_main.set_ylabel('Close Encounter Distance (au)')

# Plot histograms
ax_histx.hist(all_times, bins=50, color='blue', alpha=0.7)
ax_histx.set_yscale('log')  # Set y-axis of the top histogram to log scale

# Hide x-axis labels for the top histogram
ax_histx.set_xticklabels([])

# Ensure x-limits of both axes align correctly
ax_histx.set_xlim(ax_main.get_xlim())

plt.show()
plt.savefig("High density 100.png")
```
