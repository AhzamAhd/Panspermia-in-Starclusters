```python
import pandas as pd
import numpy as np
```

```python
data_high_10 = pd.read_csv("C:\Ahzam\MSc Astrophysics\Project\Python\data's\star_data_high_10.csv")
data_high_100 = pd.read_csv("C:\Ahzam\MSc Astrophysics\Project\Python\data's\star_data_highden_100.csv")
data_low_10 = pd.read_csv("C:\Ahzam\MSc Astrophysics\Project\Python\data's\star_data_low_10.csv")
```

```python
import matplotlib.pyplot as plt

# Filter the data for binary and non-binary stars
binary_stars_10 = data_high_10[data_high_10['binding_energy'] < 0]['close_encounter_distance'] * 206265
binary_stars_100 = data_high_100[data_high_100['binding_energy'] < 0]['close_encounter_distance'] * 206265
binary_stars_low_10 = data_low_10[data_low_10['binding_energy'] < 0]['close_encounter_distance'] * 206265
plt.figure(figsize=(10, 8))
# Create the histograms


plt.hist(binary_stars_low_10,  label='Binary Stars low density 10 Myr', color='green', edgecolor='black')
plt.hist(binary_stars_10,  label='Binary Stars high density 10 Myr', color='red',  edgecolor='black')
plt.hist(binary_stars_100,  label='Binary Stars high density 100 Myr', color='blue',  edgecolor='black')

# You may uncomment this line to use a logarithmic scale on the y-axis
# plt.yscale('log')

plt.xlabel('Close Encounter Distance (AU)')
plt.ylabel('Frequency')
plt.title('Distribution of Close Encounter Distances for Binary and Non-Binary Stars')
plt.legend()
#plt.grid(True)
plt.tight_layout()
plt.show()
# Create the histograms with custom bin numbers and alpha for better visualization

```
