```python
import pandas as pd
import numpy as np
from sklearn.neighbors import NearestNeighbors
#pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
data_1_high = pd.read_csv("C:\\Users\\Ahzam Ahmed\\Method 2 close encounter star\\high_10_e_1.csv")
data_1_low = pd.read_csv("C:\\Users\\Ahzam Ahmed\\Method 2 close encounter star\\low_10_e_1.csv")
data_3_5_low = pd.read_csv("C:\\Users\\Ahzam Ahmed\\Method 2 close encounter star\\low_10_e_3.5.csv")
data_3_5_high = pd.read_csv("C:\\Users\\Ahzam Ahmed\\Method 2 close encounter star\\high_10_e_3.5.csv")
```

```python
import matplotlib.pyplot as plt

# Sample data


plt.scatter(data_1_high['close_encounter_time'], data_1_high['rocky_materials'],label = 'High Density Environment')
plt.scatter(data_1_low['close_encounter_time'], data_1_low['rocky_materials'],label = 'Low density Environment')
plt.xlabel('Time (Myr)')
plt.ylabel('Transferred Mass in kg')
plt.yscale('log')
plt.legend(loc='upper left')
plt.show()
```

```python
plt.scatter(data_3_5_high['close_encounter_time'], data_3_5_high['rocky_materials'],label = 'High Density Environment')
plt.scatter(data_3_5_low['close_encounter_time'], data_3_5_low['rocky_materials'],label = 'Low density Environment')
plt.xlabel('Time (Myr)')
plt.ylabel('Transferred Mass in kg')
plt.yscale('log')
plt.legend(loc='upper left')
plt.show()
```

```python
bins = [0, 2, 4, 6, 8, 10]
labels = ['0-2', '2-4', '4-6', '6-8', '8-10']
data_1_high['time_interval Myr'] = pd.cut(data_1_high['close_encounter_time'], bins=bins, labels=labels, right=False)
high_grouped_1 = data_1_high.groupby('time_interval Myr')['rocky_materials'].sum()
data_3_5_high['time_interval Myr'] = pd.cut(data_3_5_high['close_encounter_time'], bins=bins, labels=labels, right=False)
high_grouped_3_5 = data_3_5_high.groupby('time_interval Myr')['rocky_materials'].sum()


data_1_low['time_interval Myr'] = pd.cut(data_1_low['close_encounter_time'], bins=bins, labels=labels, right=False)
low_grouped_1 = data_1_low.groupby('time_interval Myr')['rocky_materials'].sum()
data_3_5_low['time_interval Myr'] = pd.cut(data_3_5_low['close_encounter_time'], bins=bins, labels=labels, right=False)
low_grouped_3_5 = data_3_5_low.groupby('time_interval Myr')['rocky_materials'].sum()
table = pd.DataFrame({
    'Time interval (Myr)': labels,
    'High Density Environment, eccentricity = 1': high_grouped_1.values,
    'High Density Environment, eccentricity = 3.5': high_grouped_3_5.values,
    'Low Density Environment, eccentricity  = 1': low_grouped_1.values,
    'Low Density Environment, eccentricity  = 3.5': low_grouped_3_5.values,
})
```
