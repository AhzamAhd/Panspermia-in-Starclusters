```python
import numpy as np
import matplotlib.pyplot as plt

# Set up the parameter grids
mass_ratios = np.linspace(0.5, 2, 10) # mass ratio values
qenc_values = np.linspace(100, 1000, 100) # encounter distance values

# Create a grid of values
mass_ratio_grid, qenc_grid = np.meshgrid(mass_ratios, qenc_values)

# Calculate r_min for both eccentricities
r_min_grid_05 = 0.3 * ((1 + mass_ratio_grid) * (1 +0.5))**(-1/3) * qenc_grid
r_min_grid_35 = 0.3 * ((1 + mass_ratio_grid) * (1 + 3.5))**(-1/3) * qenc_grid

# Compute the global minimum and maximum values of r_min
global_vmin = min(np.min(r_min_grid_05), np.min(r_min_grid_35))
global_vmax = max(np.max(r_min_grid_05), np.max(r_min_grid_35))

# Plot the heatmap for eccentricity 0.5
plt.figure(figsize=(10,6))
plt.imshow(r_min_grid_05, origin='lower', extent=[mass_ratios.min(), mass_ratios.max(), qenc_values.min(), qenc_values.max()],
           aspect='auto', cmap='hot_r', vmin=global_vmin, vmax=global_vmax) # Using global vmin and vmax
plt.colorbar(label='r_min')
plt.xlabel('Mass ratio (M1/M2)')
plt.ylabel('Encounter distance (qenc)')
plt.title('Heatmap of r_min values for eccentricity 0.5')
plt.show()

# Plot the heatmap for eccentricity 3.5
plt.figure(figsize=(10,6))
plt.imshow(r_min_grid_35, origin='lower', extent=[mass_ratios.min(), mass_ratios.max(), qenc_values.min(), qenc_values.max()],
           aspect='auto', cmap='hot_r', vmin=global_vmin, vmax=global_vmax) # Using global vmin and vmax
plt.colorbar(label='r_min')
plt.xlabel('Mass ratio (M1/M2)')
plt.ylabel('Encounter distance (qenc)')
plt.title('Heatmap of r_min values for eccentricity 3.5')
plt.show()
```
